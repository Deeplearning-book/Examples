```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.rc('font',serif='Times')
```

```python
x1=np.arange(-10,10,1)
#y1=np.exp(-x1**2/10)
y1=np.zeros(20)
y1[10]=1
y1[11]=-1
x2=np.arange(-20,20,1)
y2=np.zeros(40)
y2[np.arange(10,30)]=1

x3=np.arange(-30,29)
z=np.convolve(y1,y2)
```

```python

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2)
plt.subplot(221)
markerline, stemlines, baseline =plt.stem(x1,y1,linefmt='grey', markerfmt='ko')
markerline.set_markerfacecolor('none') 
baseline.set_color('black')
plt.xlabel("$n$", fontsize=20)
plt.ylabel("$f[n]$", fontsize=20)
plt.subplot(222)
markerline, stemlines, baseline =plt.stem(x2,y2,linefmt='grey', markerfmt='ko')
markerline.set_markerfacecolor('none') 
baseline.set_color('black')
plt.xlabel("$n$", fontsize=20)
plt.ylabel("$g[n]$", fontsize=20)
plt.subplot(212)
markerline, stemlines, baseline = plt.stem(x3,z,linefmt='grey', markerfmt='ko')
markerline.set_markerfacecolor('none') 
baseline.set_color('black')

plt.xlabel("$n$", fontsize=20)
plt.ylabel("$(f*g)[n]$", fontsize=20)
fig.tight_layout()
plt.savefig("../../Pictures/Ch_3/ex_convolution_derivative.pdf")
plt.show()
```
