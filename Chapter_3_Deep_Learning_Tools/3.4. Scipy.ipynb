## **3.4.1. Data input/output using Scipy**

The __Scipy.io__ library can be used to work around with various types of file formats such as Matlab, IDL,  wav sound files, etc. The most commonly used files include _.mat_ files. The Scipy.io provides functions to load, save and work around with these _.mat_ files. Similarly, the package allows reading, writing and processing of other file formats. The following commands are useful for _.mat_ files:

1. __sio.savemat__ : Used to save data into .mat file
2. __sio.loadmat__ : This function opens and reads mat files
3. __sio.whosmat__ : This function is used to display the contents of the mat file

For each of the file formats, commands vary and these specific sets of commands can be used for reading and writing to those formats.

```python
import scipy.io as sio
import numpy as np

arr = np.array([1,2,3,4])  # create an array using np
sio.savemat('Data\sample_data.mat', {'arr_samp': arr}) #saving the array by the name 'arr_samp'
                                                  #using sio.savemat

sample_arr = sio.loadmat('Data\sample_data.mat') # loading the array into the variable from mat file
                                            # using sio.loadmat
print(sample_arr['arr_samp'])

print(sio.whosmat('Data\sample_data.mat'))
```

## **3.4.2. Clustering methods**

The __Scipy.cluster__ is used to perform clustering operation on the input data points. Clustering operation is used to divide the data into different groups. The most popular clustering approach used is _K-Means clustering_.  

The K-Means algorithm iterates the datapoints through two steps:
1. In the first step, the data points closer to the center are identified at each run using a distance measure.
2. The next step involves computing the mean of each cluster which becomes the new center.

These two steps are iterated through till the centers don't move.

_Scipy.cluster_ has the implementation of K-Means algorithm.

The following example uses the Scipy package to cluster a set of randomly generated data. The generated data are located around two different centers. Therefore, the main goal is to cluster these datasets into two different clusters using K-means algorithm.

```python
from scipy.cluster.vq import kmeans, vq, whiten #importing the packages required
#for Kmeans clustering
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('grayscale')
data1 = np.random.randn(500,2) # generating a random 2D data of normal distribution
data2 = np.random.randn(500,2) + np.array([3,3]) # adding distance to the previously generated d
# to create two clusters
data = np.vstack((data1,data2)) # stacking the two data together
[center,_] = kmeans(data, 2) # Apply Kmeans clustering to cluster the data

plt.scatter(data[:,0], data[:,1],marker = '+', color = 'k', label = 'data') # plotting the scatterplots
plt.scatter(center[:,0], center[:,1], marker = 'o', color = 'gray', s = 200, label = 'centroid') # plotting the two centroids
plt.legend()
csfont = {'fontname':'serif'} #define the font family
plt.xlabel("data1",fontsize=12,**csfont) #Set the x-axis label of the current axes
plt.ylabel("data2",fontsize=12,**csfont) #Set the y-axis label of the current axes
#plt.title("Scatterplot of the data with centroids",fontsize=12,**csfont) #Set a title of the current axes
plt.savefig("Data\kmeans.pdf")
plt.show()

# into two groups
out = vq(data, center)
print(out)
```

## **3.4.4. Linear algebra and Integration routines**

### Linear algebra
The package __scipy.linalg__ can be used to for performing linear algebra operations in python. The package is a faster implementation of BLAS and LAPACK libraries which are standard softwares used for linear algebra implementation. The package accepts two dimensional arrays and returns two dimensional arrays as outputs.

### Integration
The package __scipy.integrate__ is used for doing single, double, multiple integration. Additionally, you can use this package for applying the numerical analyisi rules such as simpson's rule, trapazoidal rule etc. These type of numerical integration becomes essential when we cannot do the integration analytically.

```python
# LINEAR ALGEBRA
from scipy import linalg
a = np.array([[1,2],[3,2]]) # create a square matrix

print('\ndeteminant of matrix a=')
print(a, linalg.det(a))    # compute the determinant of the matrix

print('\nInverse of the matrix a = ')
print(linalg.inv(a))       # compute the inverse of the matrix
val, vect = linalg.eig(a)  # calculate the eigenvalue and eigen vector of the matrix
print('\neigenvalue =')
print(val)
print('\neigenvector =')
print(vect)
b = np.array([2,4])       # create another 2D vector which is the RHS OF THE
                          # algebra equation
print('\nSolution to equation using matrix a and b gives:')
print(linalg.solve(a,b))  # print the solution to the equation using a and b

# INTEGRATION
from scipy import integrate
def f(x, y):
    return x*y
def bounds_y():
    return [0, 0.5]
def bounds_x(y):
    return [0, 1-2*y]

integ = integrate.nquad(f, [bounds_x, bounds_y])
print('\nafter integration')
print(integ)
```

## **3.4.5. Optimization**
The package used is __scipy.optimize__ is used for carrying out various basic optimizations. Basically we can define any function and use this module to optimize a parameter by computing the minimum of the function.

1. For minimization __minimize()__ function is used.
2. Singlevariate systems can be solved using functions such as __minimum_scaler()__ and __newton()__ functions.
3. For solving multivariate systems, the roots of these systems can be found using the __root()__ command.
4. The package also has most commonly used optimization approaches such as least squares __least_squares__ and curve fitting techniques __curve_fit()__.

```python
import matplotlib.pyplot as plt
from scipy import optimize
import numpy as np

def func(x):
    return   x**3 - 2*(x**2) + x + 1 # create the fucntion to be optimized

x = np.linspace(-1,2.5,num=100)    # plot the data to visualize the function
plt.plot(x, func(x))
plt.plot(1/3,func(1/3),'o', color = 'gray', markersize = 8)
plt.plot(1,func(1),'o', color = 'gray',markersize = 8)
plt.savefig("Data\Optimiz.pdf")
plt.show()


out = optimize.minimize_scalar(func) # Find the minimum of the function
                                     # using minimum_scalar
print(out)
```

## **3.4.6. Interpolation**

Scipy package __scipy.interpolate__ is used to perform this operation. If we have a function and few points to describe the sequence of the data, the interpolate function can be used to predict values that falls into the existing set of data. This can be done in 1D data. There are different kinds of interpolation : _'linear'_, _'cubic'_, _'quadratic'_ etc. Smooth curves can be plotted  through limited number of points using Splines command. We can specify the degree of the smoothing data.

```python
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt

#1D interpolation
#plt.figure(figsize=(15,10),dpi=90)
csfont = {'fontname':'serif'} #define the font family
x = np.linspace(0, 10, 10)
y = np.cos(2*x) + 1          # creating the input signal with 10 points
#plt.subplot(2,3,1)
plt.plot(x, y,'o')
plt.legend(['input signal']) # plotting the input signal points
plt.xticks(fontsize=12,**csfont)
plt.yticks(fontsize=12,**csfont)
plt.savefig('Data\input_interp.pdf')
plt.show()

f1 = interpolate.interp1d(x, y,kind = 'linear') # performing 1D linear interpolation
f2 = interpolate.interp1d(x, y, kind = 'cubic') # performing 1D cubic interpolation
f3 = interpolate.interp1d(x, y, kind = 'quadratic') # performing 1D quadratic interpolation

x_new = np.linspace(0, 10, 30)
#plt.subplot(2,3,2)
plt.plot(x, y, 'o', x_new, f1(x_new), '-', x_new, f2(x_new), '--', x_new, f3(x_new), '-*')
                                                # plotting all the interpolatid data points
                                                # alongwith the original data
plt.legend(['data', 'linear', 'cubic','quadratic'])
#plt.title('1D interpolation')
plt.xticks(fontsize=12,**csfont)
plt.yticks(fontsize=12,**csfont)
plt.savefig('Data\interp_1D.pdf')
plt.show()

# using spline

x = np.linspace(0, 6, 50) # creating the input signal
y = x**2 + np.sin(x)
n = np.random.randn(50) # adding random noise to the data to
data = y + n            # make the signal distorted
#plt.subplot(2,3,4)
plt.plot(x, data, 'o') # plotting the generated data
plt.xticks(fontsize=12,**csfont)
plt.yticks(fontsize=12,**csfont)
#plt.title('input signal')
plt.savefig('Data\input2_interp.pdf')
plt.show()

smth = interpolate.UnivariateSpline(x, data) # performing interpolation
                                             # and smoothing using spline
x_new = np.linspace(0, 6, 1000)
#plt.subplot(2,3,5)
plt.plot(x, data, 'o', x_new, smth(x_new), '-')
plt.legend(['data', 'smooth curve'])
plt.xticks(fontsize=12,**csfont)
plt.yticks(fontsize=12,**csfont)
#plt.title('smoothing and curve fitting using spline package')
plt.savefig('Data\interp_smooth.pdf')
plt.show()
smth.set_smoothing_factor(0.5)    # adjusting the smoothing factor
#plt.subplot(2,3,6)
plt.plot(x, data, 'o', x_new, smth(x_new),'-')
plt.legend(['data', 'smooth curve'])
plt.xticks(fontsize=12,**csfont)
plt.yticks(fontsize=12,**csfont)
#plt.title('after setting the smooth factor as 0.5')
plt.savefig('Data\interpolation.pdf')
plt.show()


```

## **3.4.7 Image Processing**

The package __scipy.ndimage__ can be used to perform image processing operations in python. The package can be used to:
1. Perform geometric transformations such as rotation etc.
2. image filtering
3. Edge detection

This package is a powerful tool to perform image manipulation. The scipy package also has access image through a library called \emph{MISC} that can be used as a sample to perform the operations.

## **Example 3.4.1 (mage Processing Operations using Scipy package)**

In this example, we are going to learn how to display an image imported from MISCpackage and apply image processing techniques to this image. Import the faceimage using the misc library in scipy. Display the image and now let us performsome geometric transformations to the image. The ndimage library allows usto do several operations such as rotation, flipping and cropping. Let us performa downward flip using numpy package followed by a rotation by an angle of 60degrees using ndimage library. In addition to geometric transformation we canalso do several filtering operations on the images. Begin with blurring the imageusing a standard gaussian filter. Different kinds of filters can be accessed usingthe ndimage. Further, the sigma value can be tuned to see the difference betweenthe intensity of blurring. Another important image processing that can be per-formed using this package is the edge detection. Several methods are available.Let us apply the Canny and the Prewitt algorithm to the original image and thecropped image. Display all the output images

```python
from scipy import misc
import numpy as np
from scipy import ndimage
import matplotlib.pyplot as plt

image = misc.face()  # loading the image
plt.imshow(image)
plt.show()

flip_im = np.flipud(image) # flip downward the image
rot_im = ndimage.rotate(flip_im, 60)
plt.imshow(flip_im)
plt.title('flipped image')
plt.show()
plt.imshow(rot_im)        # rotate the image by 60 degrees
plt.title('rotated image')
plt.show()

blurred = ndimage.gaussian_filter(image, sigma=5) # perform gaussian blur
plt.imshow(blurred)
plt.title('blurred image')
plt.show()

blurred1 = ndimage.gaussian_filter(image, sigma=2) # perform gaussian blur
                                                   # with sigma=2
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Gaussian blurred image with sigma=5 vs sigma=2') # plotting the
ax1.imshow(blurred)                                           # two graphs
ax1.set_title('sigma=5')
ax2.imshow(blurred1)
ax2.set_title('sigma=2')
plt.show()

sx = ndimage.prewitt(image[:,:,0])    # taking one of the channels of the image
                                      # and applying the edge detection to it
plt.imshow(sx)
plt.title('edge detection using Prewitt algorithm')
plt.show()
```

## **3.4.8. Special functions**

Apart from all the above mentioned libraries, Scipy also provides us with special functions. __scipy.special__ package give us access to certain universal functions sucha s cubic root, exponential, Permutation and combinations, gamma functions etc. The syntax for each of these functions are slightly different. In case of mathematical functions, i.e., exponential, gamma, cubic root,  element wise operations are performed.

```python
import numpy as np
from scipy import special

a = np.array([1, 3, 5, 2])

print(special.cbrt(a)) # computing the cube root of a
print(special.exp10(a))# computing the 10^a
print(special.gamma(a))# computing the gamma to s

out_comb = special.comb(5, 2, exact = False,repetition=True)
                    # Combination(5,2)
print(out_comb)
out_perm = special.perm(5, 2, exact = True)
print(out_perm)     #permutation(5,2)
```
