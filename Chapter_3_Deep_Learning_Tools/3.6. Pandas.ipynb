## **3.6.3 PANDAS DATA MANIPULATION**
### **Sorting**
- It can be done in two ways,
1. Sorting by label: The sort_index() method is used to sort row labels of the dataframe and the default sorting is done in ascending order. To sort by column labels, the axis argument of the sort_index() function  is set to 1.
2. Sorting by value: The sort_values() method of pandas is used to sort the values of the dataframe along either axis by defining a ‘by’ argument.



```python
import pandas as pd
import numpy as np
list_list = [['Canada','Ottawa'],['Italy','Rome'],['Peru','Lima']] #define a list of lists.
df_nosort = pd.DataFrame(list_list,index=[3,1,2],columns=['Country','Capital']) #create the dataframe.
print(df_nosort) #outputs unsorted dataframe.
df_sorta = df_nosort.sort_index() #sorting by label in ascending order.
df_sortd = df_nosort.sort_index(ascending=False) #sorting by label in descending order.
df_sortc = df_nosort.sort_index(axis=1) #sorting by column label.
print(df_sorta) #outputs sorted dataframe.
print(df_sortd) #outputs sorted dataframe.
print(df_sortc) #outputs sorted dataframe.
```

```python
import pandas as pd
datax = {'A': [10,40,30,25],'B': [2016,2007,2020,2001]} #define a dictionary of lists.
dfx_unsort = pd.DataFrame(datax, columns=['A','B']) #create a dataframe.
print(dfx_unsort) #outputs unsorted dataframe.
dfx_sort = dfx_unsort.sort_values(by = 'B') #sort by values along an axis.
dfx_sortm = dfx_unsort.sort_values(by = ['A','B']) #sort by multiple columns.
print(dfx_sortm) #outputs sorted dataframe.
print(dfx_sortm.rank()) #assign rank to the values of the dataframe and outputs the result.
```

### **Grouping**
- This operation is used to split the data into groups and apply a function to these subsets.
The most common functions are as follows:
1. Aggregation: It can be used to provide a statistical summary by computing the mean, sum, minimum, maximum, etc.  
2. Transformation: A specific operation is performed so that the output and input data frames are able to maintain the same size.
3. Filtration: It is used to eliminate unwanted data using a condition.

```python
import pandas as pd

data1 = {'Date':['1-1-2021', '1-2-2021', '1-3-2021', '1-4-2021',
                 '1-5-2021', '1-6-2021', '1-7-2021', '1-8-2021'],
        'High':[270, 214, 522, 632,
               343, 836, 427, 132],
        'Low':[16, 24, 22, 32,
               33, 36, 27, 32],
        'Open':[45,183,261,382,
                90,534,60,55],
         'Close':[103,89,333, 384,
               200,136,68,118]} #define a dictionary of lists.
df = pd.DataFrame(data1) #create a dataframe
print(df)
group1 = df.groupby('Date')
print(df.groupby('Date').groups) #grouping data with one key
print(group1.get_group('1-6-2021')) #selects a single group
```

```python
import pandas as pd
import numpy as np
dataw = {'Name':['Fred', 'George', 'Fred', 'Harry', 'Ron', 'Percy', 'Harry', 'Fred'], #define a dictionary of lists.
        'Marks':[80, 95, 75, 70, 98, 60, 75, 62]}
df = pd.DataFrame(dataw) #create a dataframe.
print(df)
groupw = df.groupby('Name') #grouping data with one key.
print(groupw.aggregate(np.sum)) #performing aggregation using aggregate method.
tf = lambda x: x - x.mean() #subtracting the mean.
print(groupw.transform(tf))  #applying the transform function.
print(groupw.filter(lambda x: len(x) >= 2)) #filtering the data.
```

## **3.6.4 HANDLING MISSING DATA**
- This is a common problem prevalent in most real-time databases.
- The tabulated data might have multiple missing values leading to reduced performance for many data science problems.
- Henceforth data cleaning is highly essential and one of the most powerful tools employed is pandas.
- The standard missing values in pandas are NaN( Not a number), NA(Not available), and None.
- In order to detect the missing values, pandas use the isnull() and nonnull() functions. They output the boolean values to denote the missing values in the dataframe.

```python
import pandas as pd
import numpy as np
datag = {'A':[None, 90, np.nan, 95],
        'B': [30, 45, 56, np.nan],
        'C':[np.nan, 40, 80, None],
        'D':[70, 40, 24, None]} #define a dictionary of lists.
df = pd.DataFrame(datag) #create a dataframe.
print(df.isnull()) #check the missing values using isnull().
print(df.notnull()) #check the missing values using notnull().
```

After finding the missing values, it is possible to either drop or replaces these values in the dataframe.
1. The dropna() function is used to drop a row or column with null values. It has three parameters axis, how and thresh. The default value of the axis is 0 representing the rows whereas the how parameter takes either of the two string values ‘any’ or ‘all’. The third parameter thresh inputs an integer that corresponds to a threshold for missing values in order for a row or column to be dropped.
2. The fillna() function on the other hand is used to replace the missing values.

```python
import pandas as pd
listq = [[200,np.nan,800],[np.nan,np.nan,600],
         [900,None,None],[500,np.nan,None]] #define a list of lists.
df = pd.DataFrame(listq,columns=['Column1','Column2','Column3']) #create the dataframe.
df.dropna(axis=1,how='any')
print(df)
df1=df.dropna(axis=1,how='all') #drops a column when all of its values are missing.
df2=df.dropna(axis=1,thresh=2) #drops a column with atleast 2 missing values.
df3=df.fillna(0) #replaces the missing values with zeros.
print(df1)
print(df2)
print(df3)
```

## **3.6.5 INPUT-OUTPUT TOOLS**

- In practical problems, we usually work with data that already exists.
- This data is saved in multiple formats and it is most commonly stored as comma-separated values also known as csv files.
- Pandas use the read_csv function to read a dataframe and to_csv function to save the dataframe as a csv file


```python
import numpy as np

class Data:
    def __init__(self, N, sigma):
        self.N=N #Number of data per cluster.
        self.sigma=sigma #standard deviation of the clusters.

    def data_xor(self, classes):
        X=self.sigma*np.random.randn(2,4*self.N) # Generate 4*N random vectors (gaussian) centered around zero
        mean=np.array([[-1,-1, 1, 1],[-1,1,-1,1]]) # define four means
        M=np.ones((self.N,2))*mean[:,0] # Means of the first cluster
        y=np.ones((1,self.N))*classes[0] # Labels of the first cluster
        for i in range(1,4):
            m=np.ones((self.N,2))*mean[:,i] # Means of cluster i
            M=np.concatenate((M,m))     #Concatenate all means
            y=np.concatenate((y,np.ones((1,self.N))*classes[i]),axis=1) # concatenate labels
        M=M.T
        X=X+M # Add means to the data.
        return X,y

    def data_circle(self):
        theta=np.random.rand(1,self.N)*np.pi*2
        rho=np.random.randn(1,self.N)*self.sigma+1
        X1=rho*np.block([[np.cos(theta)],[np.sin(theta)]]) #circular data corresponding to class 1

        theta=np.random.rand(1,self.N)*2*np.pi
        rho=np.random.randn(1,self.N)*self.sigma+0.8
        X2=rho*np.block([[np.cos(theta)],[np.sin(theta)]])

        y=np.concatenate((0*np.ones((1,self.N)),np.ones((1,self.N))),axis=1) # labels
        X=np.concatenate((X1,X2),axis=1)
        return X,y
#The class definition ends here.
```

```python
np.random.seed(30) #allows to reproduce the same results.
N=100; sigma=0.6 #attributes for xor.
classes=[0,1,1,0] #attribute for xor.
T=Data(N,sigma) #define the object corresponding to xor data.
X,y=T.data_xor(classes) #method with instance variable classes.
print(X.shape,y.shape)
```

```python
import pandas as pd
#create a dictionary and define the dataframe.
df = pd.DataFrame({'Feature1':X[0,:],'Feature2':X[1,:],'Label': y[0,:]}, columns=['Feature1','Feature2','Label'])
print(df)
df.to_csv('Data\pandas_write.csv',index=False) #write to a csv file and drop the index.
```

## **3.6.7 DATA OPERATIONS**

- Data analysis in pandas is further facilitated using functions that perform the merge, join, append and concatenate operations.
- Both merge() and join() functions supports horizontal combination whereas append() and concat() are used for vertical combination. The main role of merge() function is to combine the dataframe with common columns or indices.
- The different types of merge operations performed are:
 1. left: use only keys from the left dataframe while preserving the key order.
 2. right: use only keys from the right dataframe while preserving the key order.
 3. inner: uses the intersection of keys from both left and right dataframe while preserving the order of only left keys.
 4. outer: uses the intersection of keys from both left and right dataframe while sorting the keys lexicographically.



```python
import pandas as pd
df_left=pd.DataFrame({"keyx": ["A", "B", "C", "D"],
        "keyy": ["A", "A", "D", "B"],
        "Name": ["Alex", "Daniella", "Fiona", "Kiara"],
        "Age": ["23", "31", "26", "29"]}) #define a dataframe.
df_right = pd.DataFrame({"keyx": ["A", "B", "C", "D"],
        "keyy": ["A", "A", "A", "A"],
        "Job": ["Writer", "Painter", "Doctor", "Engineer"],
        "Country": ["Russia", "France", "Italy", "Greece"]}) #define another dataframe.
out = pd.merge(df_left, df_right, on=["keyx", "keyy"]) #merge two dataframes using the keys.
out1 = pd.merge(df_left, df_right, how='outer') #merges the union of keys from both dataframes.
print(df_left) #outputs the given dataframe.
print(df_right) #outputs the given dataframe.
print(out) #outputs the merged dataframe.
print(out1) #outputs the merged dataframe.
```

- The join() function is used to combine the columns of two dataframes with different indexes.
- Here only one dataframe can be specified and later by using the join() function it can be combined with the other dataframe.
- This function uses the lsuffix and rsuffix parameters to define a suffix to add to any overlapping column.



```python
import pandas as pd
left_df = pd.DataFrame({"X": ["X1", "X2", "X3", "X4"], "Y": ["Y1", "Y2", "Y3", "Y4"]}, index=[1,2,3,4]) #create a dataframe.
right_df = pd.DataFrame({"X": ["X11", "X12", "X13"], "Q": ["Q1", "Q2", "Q3"]}, index=[1,3,4]) #create another dataframe.
out = left_df.join(right_df,lsuffix='_left', rsuffix='_right') #join the dataframe with overlapping columns.
print(out) #outputs the join dataframe.
```

- The concat() function links the two dataframes along an axis, either rows or columns.
- The append() function is used as a simple yet efficient shortcut for concatenation.

```python
import pandas as pd
dfx = pd.DataFrame({'Name': ['A', 'B', 'C', 'D'],
    'Marks': [50,99,72,60]}) #define a dataframe.
dfy = pd.DataFrame({'Name': ['E', 'F', 'G', 'H'],
    'Marks': [76,85,93,96]}) #define another dataframe.
out = pd.concat([dfx, dfy]) #concatenate along the rows.
out1 = pd.concat([dfx, dfy], axis=1) #concatenate along the columns.
out2 = pd.concat([dfx, dfy], ignore_index=True) #concatenate to ignore existing indices.
#out3 = dfy.append(dfx) #append along the rows.#Here append is deprecated.
out3 = pd.concat([dfy, dfx]) # Use pd.concat instead of append
print(out) #outputs the concatenated dataframe.
print(out1) #outputs the concatenated dataframe.
print(out2) #outputs the concatenated dataframe.
print(out3) #outputs the appended dataframe.
```

## **3.6.8 DATA VISUALIZATION**
- Pandas provide an easy-to-use interface built over matplotlib for data visualization by using less code.
- Pandas visualization tool uses the plot() function to produce
 1. line plots
 2. scatter plots
 3. bar graphs
 4. histograms
 5. pie plots
 6. area plots
 7. box plots
 8. kernel density estimation plots.
- It essentially provides a graphical representation of all the columns along with their labels.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('grayscale')
#define a dataframe.
df = pd.DataFrame(np.random.rand(5,3),columns=('Data1', 'Data2', 'Data3'))
print(df) #outputs the dataframe.
df.plot() #line plot
df.plot(subplots=True, figsize=(10,10)) #create subplots for each columns.
#scatter plot to check correlation.
df.plot(x="Data1", y="Data2", kind="scatter",color="black", marker="*", s=100)
```

```python
import matplotlib.pyplot as plt
plt.style.use('grayscale')
fig, axes = plt.subplots(nrows=1, ncols=3,figsize=(15, 5))
df.plot(kind="bar",ax=axes[0]) #bar plot.
#stacked bar plot to highlight the comparison between different categories.
df.plot(kind="bar", stacked=True,grid = True, legend = True,ax=axes[1])
df.plot(kind="barh", stacked=True,grid = True, legend = True,ax=axes[2]) #horizontally stacked bar plot.
plt.savefig('Data\pandas_bar.pdf')
plt.show()
```
