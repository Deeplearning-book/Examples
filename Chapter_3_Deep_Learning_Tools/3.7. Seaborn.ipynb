## **3.7.2 PLOTTING WITH SEABORN**
### **UNIVARIATE PLOTS**
- These plots are used to describe the nature of a single variable and it represents the rate of occurence of each uniques values of a given variable.
- These visualizations are done using barplots and histograms and the commonly used functions are the histplot(), distplot() and countplot().
- A histplot() outputs a histogram by plotting the distribution of the given variable.
- A distplot() also represents a histogram with a continous function passed on top of the bins in the form of a line by using Kernel Density Estimate.
- In case of countplot() function, it is used to display the distribution of categorical variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('iris') #loading dataset.
#set figure
f, ax = plt.subplots(1,1)
csfont = {'fontname':'serif'} #define the font family
#graph histogram
sns.histplot(df['sepal_length'], bins=10, alpha=0.5, color='black', label='sepal length')
sns.histplot(df['sepal_width'], bins=10, alpha = 0.5, color='grey', label='sepal width')
#set legend
plt.legend(loc='upper right')
#set title & axis titles
plt.xlabel('Frequency',fontsize=12,**csfont) #labels the x-axis.
plt.ylabel('Values',fontsize=12,**csfont) #labels the y-axis.
plt.xticks(fontsize=12,**csfont)
plt.yticks(fontsize=12,**csfont)
#set x & y ranges
plt.xlim(0,10)
plt.ylim(0,40)
plt.savefig('Data/uni.pdf')
plt.show()
```

```python
sns.countplot(x="species", data=df)
plt.title("Count of Species")
plt.show()
```

### **BIVARIATE PLOTS**
- These plot are mainly used to demonstrate the relationaship between two variables.
- Depending upon the type of variables, different types of functions such as scatterplot(), lineplot() and jointplot() are used for plotting.
- Scatterplot() provides a joint representation of two or more variables by using a set of points
- It is available in both 2D as well as 3D.
- Lineplot() is a similar version of scatterplot() where the data is represented using a line.
- Jointplots() function is used in both univariate and bivariate plotting.
- The central plot will correspond to a bivariate analysis, whereas the univariate plots of the variables are displayed on the top and right side of the graph.


```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('grayscale')
df1 = sns.load_dataset('tips') #loading dataset.
df1.head() #viewing the first 5 rows of the dataframe.
```

```python
sns.scatterplot(x='total_bill', y='tip', data=df1, hue='time')
plt.title("Scatterplot of total_bill vs. tip") #displays the title.
plt.show()
```

```python
sns.lineplot(x="total_bill", y="tip", data=df1)
plt.title("Lineplot of total_bill vs. tip") #displays the title.
plt.show()
```

```python
h=sns.jointplot(x='total_bill', y='tip', data=df1)
csfont = {'fontname':'serif'} #define the font family
h.ax_joint.set_xlabel('total_bill',fontsize=12,**csfont) #custom axis label for x axis
h.ax_joint.set_ylabel('tip',fontsize=12,**csfont) #custom axis label for y axis
plt.savefig('Data\joint.pdf')
plt.show()
```

### **MULTIVARIATE PLOTS**
- These plots can show the relationship between three or more different variables.
- Inorder to separate features in multiple dimensions, seaborn uses the hue parameter.
- The different plot functions used for multivariate analysis are scatterplot(), barplot() and pairplot().
- While plotting three variables in scatterplot(), two variables can be numerical variables while the third variable can be categorical and it can be used differentiate the groups within the data.
- Inorder to aggregrate categorical data, we use barplot() to illustrate the relationship between the variables.
- Here two variables can be categorical wheras the third can be a numerical variable.
- Finally the pairplot() function shows the pairwise relationships in a dataset.


```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('grayscale')
df2 = sns.load_dataset('titanic') #loading dataset.
df2.head() #viewing the first 5 rows of the dataframe.
```

```python
sns.scatterplot(x='age', y='fare', data=df2, hue='class')
csfont = {'fontname':'serif'} #define the font family
plt.xlabel('age',fontsize=12,**csfont)
plt.ylabel('fare',fontsize=12,**csfont)
plt.savefig('Data/multi.pdf')
plt.show()
```

```python
sns.barplot(x='who', y='age', data=df2, hue='class')
plt.show()
```

```python
sns.pairplot(df2[['survived', 'pclass', 'age', 'sibsp', 'parch']], hue='survived')
plt.show()
```

## **3.7.3 ADDITIONAL PLOTTING FUNCTIONS**
### **Correlation Plots**
- In regression models, it is essential to check the correlation between the variables.
- Seaborn uses two main functions, regplot() and lmplot() to visualize this linear relationship.
- Both these functions create a scatterplot of two variables, fit the regression model and also plot the resulting regression line.
- In each case the resulting plots are identical, except that the shapes of the figures are different.



```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('grayscale')
df5 = sns.load_dataset('penguins') #loading dataset.
```

```python
sns.regplot(x="bill_length_mm", y="bill_depth_mm", data=df5)
plt.show()
```

```python
sns.lmplot(x="bill_length_mm", y="bill_depth_mm", data=df5, order=1)
csfont = {'fontname':'serif'} #define the font family
plt.xlabel('bill_length_mm',fontsize=12,**csfont)
plt.ylabel('bill_depth_mm',fontsize=12,**csfont)
plt.savefig('Data/lm.pdf')
plt.show()
```

### **Point Plots**
- These plots helps to visualize the show point estimates and confidence intervals of values at each level of the categorical variable.
- They are mainly used to find the change in a variable as it shows the mean values and error rate surrounding those mean values.

```python
sns.pointplot(x="sex", y="body_mass_g", hue="species", data=df5)
plt.show()
```

### **Cat Plots**
- Catplot shows the frequencies of the classes of one or more categorical variables.
- The catplot() function provides a new framework to visualize and compare different features of the data by giving access to several types of plots.
- Currently, catplot encompasses 8 different categorical plots available in Seaborn.
- We are able to specify the type of plot using the kind parameter and the default value for it is "strip", denoting to stripplot().


```python
sns.catplot(x="sex", y="body_mass_g", hue="species", data=df5)
csfont = {'fontname':'serif'} #define the font family
plt.xlabel('sex',fontsize=12,**csfont)
plt.ylabel('body_mass_g',fontsize=12,**csfont)
plt.savefig('Data/cat.pdf')
plt.show()
```
